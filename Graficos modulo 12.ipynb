```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dados fornecidos
data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profisso': ['Engenheiro', 'Mdico', 'Professor', 'Advogado', 'Mdico', 'Engenheiro', 'Estudante', 'Estudante'],
    'Salrio': ['4500', '8000', '5000', '10000', '12000', '15000', '1200', '1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000', '2000', '500', '250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0', '1', '0', '1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento', 'Solteiro', 'Solteiro', 'Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1', '1', '0', '0']
}

# Criar DataFrame
df = pd.DataFrame(data)

# Converso dos valores de Salrio e Limite de Crdito para numricos
df['Salrio'] = pd.to_numeric(df['Salrio'])
df['Limite_Credito'] = pd.to_numeric(df['Limite_Credito'])

# Grfico 1: Scatter Plot (Salrio vs Limite de Crdito)
plt.figure(figsize=(8, 6))
plt.scatter(df['Salrio'], df['Limite_Credito'], color='blue')
plt.title('Salrio vs Limite de Crdito')
plt.xlabel('Salrio')
plt.ylabel('Limite de Crdito')
plt.grid(True)
plt.show()

# Grfico 2: Boxplot (Estado Civil vs Limite de Crdito)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Estado_Civil', y='Limite_Credito', data=df)
plt.title('Distribuio do Limite de Crdito por Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Limite de Crdito')
plt.show()

# Grfico 3: Grfico de Barras (Imvel Prprio vs Limite de Crdito)
df_grouped = df.groupby('Imovel_Proprio')['Limite_Credito'].mean()

plt.figure(figsize=(8, 6))
df_grouped.plot(kind='bar', color='green')
plt.title('Mdia do Limite de Crdito por Propriedade de Imvel')
plt.xlabel('Imvel Prprio (0 = No, 1 = Sim)')
plt.ylabel('Mdia do Limite de Crdito')
plt.show()
```
